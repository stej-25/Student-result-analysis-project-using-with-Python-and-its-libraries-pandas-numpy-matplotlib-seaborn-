```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("student_scores.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

**Drop unnamed column**

```python
df = df.drop("Unnamed: 0",axis = 1)
print(df.head())
```

```python
df.head()
```

**gender distribution**

```python
plt.figure(figsize = (5,5))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
#from the above chart we have analysed that:
#the number of females in the data is more than the number of males
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean', "ReadingScore":"mean", "WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize= (4,4))
sns.heatmap(gb,  annot=True)
plt.title("Relationship between Parent's Education and Student's Score")
plt.show()
```

```python
#from the above chart we have concluded that the education of the parents have a good impact on their scores
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean', "ReadingScore":"mean", "WritingScore":"mean"})
print(gb1)
```

```python
plt.figure(figsize= (4,4))
sns.heatmap(gb1,  annot=True)
plt.title("Relationship between Parent's Marital Status and Student's Score")
plt.show()
```

```python
#from the above chart we have concluded that there is no\negligible impact on the
#student's score due to their parent's marital status
```

**Distribution of Ethnic Groups**

```python
# Create a series of counts for each ethnic group
ethnic_counts = df['EthnicGroup'].value_counts()

# Plot the counts as a pie chart
ethnic_counts.plot(
    kind='pie',
    autopct='%1.1f%%',
    title='Distribution of Ethnic Groups'
)

plt.ylabel('')  # Optional: clean up the y-label
plt.show()
```
